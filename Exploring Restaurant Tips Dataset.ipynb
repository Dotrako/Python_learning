## Tips Dataset

- **Size**: The dataset contains 244 rows and multiple columns.
- **Columns**: Key columns include 'total_bill,' 'tip,' 'sex,' 'smoker,' 'day,' 'time,' and 'size.'
- **Analysis Potential**: Suitable for exploring bill and tip distributions, identifying trends by day and time, and studying tipping behavior.
- **Visualization**: Offers opportunities for bar plots, scatter plots, and more.
- **Real-World Relevance**: Represents real restaurant data for learning data analysis and making informed business decisions.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the 'tips' dataset from Seaborn
tips = sns.load_dataset('tips')
```

```python
# Display the first few rows of the dataset
print(tips.head())
```

```python
# Summary statistics of the dataset
print(tips.describe())
```

```python
# Create a simple bar plot using Seaborn
sns.set(style="whitegrid")
sns.barplot(x="day", y="total_bill", data=tips, ci=None)
plt.title("Average Total Bill by Day")
plt.show()
```

```python
# Create a scatter plot to explore the relationship between 'total_bill' and 'tip'
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.show()
```
